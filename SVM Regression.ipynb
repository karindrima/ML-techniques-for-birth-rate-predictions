```python
#loading the dataset
import pandas as pd
mydata = pd.read_csv('C:\\Users\\HPP\\Desktop\\Demo Data.csv')
```

```python
mydata.describe()
```

```python
mydata.corr()
```

```python
import numpy as np
NAs = pd.concat([mydata.isnull().sum()], axis=1, keys=["mydata"])
NAs[NAs.sum(axis=1) > 0]
```

```python
#filling missing values with mean
mydata["Sex_of_HH"] = mydata["Sex_of_HH"].fillna(mydata["Sex_of_HH"].mean())
mydata.corr()
```

```python
#input and output features
X = mydata.drop(['Region', 'TCEB'], axis = 1)
X.shape
```

```python
y = mydata['TCEB']
y.shape
```

```python
#splitting into training and testing datasets
#importing train_test_split function
from sklearn.model_selection import train_test_split
#splitting the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 30)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
#support vector regression
from sklearn.svm import SVR
```

```python
#creating and fitting the SVR model
LL_svr = SVR().fit(X_train, y_train) #need to choose kernel and put in SVR()
```

```python
print('The r_sq is: %.2f'% LL_svr.score(X_train, y_train))
```

```python
#training model evaluation
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, explained_variance_score
```

```python
#predicting on training data
ytrain_pred = LL_svr.predict(X_train)
print('The r_sq is %.2f'% r2_score(y_train, ytrain_pred))
print('The Mean absolute error is: %.2f'% mean_absolute_error(y_train, ytrain_pred))
print('The Mean sq error is: %.2f'% mean_squared_error(y_train, ytrain_pred))
print('The root mean square error is: %.2f'% np.sqrt(mean_squared_error(y_train, ytrain_pred)))
```

```python
#the explained variance score to compare
print('The EVS is: %.2f'% explained_variance_score(y_train, ytrain_pred))
```

```python
#prediction on the testing data
ytest_pred=LL_svr.predict(X_test)
print(ytest_pred[:10])
```

```python
#the coefficient of determination
print('The r_sq is %.2f'% r2_score(y_test, ytest_pred))
print('The Mean absolute error is: %.2f'% mean_absolute_error(y_test, ytest_pred))
print('The Mean sq error is: %.2f'% mean_squared_error(y_test, ytest_pred))
print('The root mean square error is: %.2f'% np.sqrt(mean_squared_error(y_test, ytest_pred)))
```

```python
#plotting the observed and predicted data
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (10, 6)
x_ax = range(len(X_test))
plt.plot(x_ax, y_test, label = 'Observed', color = 'k', linestyle = '-')
plt.plot(x_ax, ytest_pred, label = 'Oredicted', color = 'k', linestyle = '--')
plt.ylabel('TCEB')
plt.xlabel('Testing sample data')
```
